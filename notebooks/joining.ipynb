```python
# import libraries
import pandas as pd
import numpy as np

# load the two tables
table1 = pd.read_excel('../data/processed/reviews_cleaned.xlsx')
table2 = pd.read_excel('../data/processed/banks.xlsx')
```

```python
# join the tables on the bank_id column
joined = table1.merge(table2, on='bank_id')
joined.head()
```

```python
# export the joined table as a excel file
joined.to_excel('../data/processed/dataset.xlsx', index=False)
```
